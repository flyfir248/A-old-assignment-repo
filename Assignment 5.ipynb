## b) 
Assume that data is a pandas DataFrame holding the breast cancer dataset. Explain in your own
words, what the following commands accomplish.

**Explaination : a.)** 

The command **"df.loc[:,'thickness']"** displays the column values of the highlighted column index "thickness" from the
dataframe df, it shows the label cross section.

```python
# Demo : a) data.loc[:,’thickness’] 

df.loc[:,'thickness'] 
```

**Explaination : b.)** 

The command **"df.loc[:,['thickness', 'mitoses']]"** displays the column values "thickness" and "mitoses" from the
dataframe df, it shows multiple label cross section.

```python
# Demo : b) data.loc[:,[’thickness’, ’mitoses’]]

df.loc[:,['thickness', 'mitoses']]
```

**Explaination : c.)** 

The command **"df.loc[2:5, :]"** displays all the column values from rows 2 to 5 of the dataframe df, it also shows multiple label cross section however this time across the entire table.

```python
# Demo : c) data.loc[2:5, :] 

df.loc[2:5, :] 
```

**Explaination : d.)** 

The command **"df.loc[:,’class’]==2 "** returns True boolean values for all the rows where column 'class' has value equal to '2', it also returns boolean false if this criteria set here is not met.

```python
# Demo : d) data.loc[:,’class’]==2

df.loc[:,'class']==2 
```

**Explaination : e.)** 

The command **"df.loc[data.loc[:,’class’]==2, [’thickness’, ’mitoses’]]"** returns only the columns thickness and mitoses where the column 'class' value matches '2'. This is evident from 458 rows returned which have been listed as benign in 'breast-cancer-wisconsin.names'. So we retieved **'Clump Thickness' & 'Mitoses' for the benign cancer(Class 2)**.

```python
# Demo : e) data.loc[data.loc[:,’class’]==2, [’thickness’, ’mitoses’]]

df.loc[df.loc[:,'class']==2, ['thickness', 'mitoses']]
```

**Explaination : f.)** 

The command **"df.iloc[:, [0, -2, -1]]"** returns only the columns which have been selected by index values, here the column headers code mitoses and class are selected.

```python
# Demo : f) data.iloc[:, [0, -2, -1]]

df.iloc[:, [0, -2, -1]]
```

```python
df.head()   # code index value 0, while mitosis and class have index values -1 and -2
```

**Explaination : g.)** 

The command **"df.iloc[:5,[0, -2, -1]]"** returns only the 1st 5 index columns (from 0 to 4) which have been selected as position of the passed integers, here the column headers code mitoses and class are selected, but only the rows from 0 till 4 have been shown.

```python
# Demo : g) data.iloc[:5,[0, -2, -1]]

df.iloc[:5,[0, -2, -1]]
```

**Explaination : h.)** 

The command **"df[numpy.logical_and(data.thickness > 3, data.epiCelSize<7)]"** returns only the rows from the dataframe which have 'Clump Thickness' greater than 3 and 'Single Epithelial Cell Size' less than 7. Here np.logical_and computes the truth value of the two columns passed.

```python
# Demo : h) data[numpy.logical_and(data.thickness > 3, data.epiCelSize<7)]

df[np.logical_and(df.thickness > 3, df.epiCelSize<7)]
```

**Explaination : i.)** 

The command **"df.loc[:, [’mitoses’, ’thickness’]].iloc[:100, :]"** returns only the 1st 100 rows via as a cross section for the labels mitoses and thickness, this is due to the positional value having been passed as 100.

```python
# Demo : i) data.loc[:, [’mitoses’, ’thickness’]].iloc[:100, :]

df.loc[:, ['mitoses', 'thickness']].iloc[:100, :]
```

## f) 
pandas functions often have a parameter axis which can be set to 0 or 1. Which axis is the
row-axis and which axis is the column-axis? (1P)

NumPy axes are very similar to axes in a Cartesian coordinate system.

Axis 0 is direction along the rows i.e. the first axis or the one which runs downward down the rows.

So to summarize **axis 0 is to go along the columns and across the rows.**

Axis 1 is direction along the columns i.e. the second axis or the one which runs right towards the columns.

So to summarize **axis 1 is to go along the rows and across the columns.**
